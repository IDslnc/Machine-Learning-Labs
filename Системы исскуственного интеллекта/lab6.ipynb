В файле water.txt представлено 61 наблюдение.
Каждое наблюдение – город в Англии и Уэльсе.
Города поделены на северные и южные.
Для каждого города известны средняя годовая смертность на 100000
населения (по данным 1958–1964) и концентрация кальция в питьевой воде (в
частях на миллион).
Чем выше концентрация кальция, тем жёстче вода.

```python
import pandas as pd
from scipy import stats
data = pd.read_csv('water.txt', delimiter='\t')
data
```

1) Используя метод .describe() вычислите описательные статистики для
северных и южных городов. Сравните средние значения смертности в
северных и южных городах и значения концентрации кальция в питьевой
воде.

```python
# Делим данные на северные и южные города
north_cities = data[data['location'] == 'North']
south_cities = data[data['location'] == 'South']

# Вычисляем описательные статистики
north_stats = north_cities[['mortality', 'hardness']].describe()
south_stats = south_cities[['mortality', 'hardness']].describe()

print("Описательные статистики для северных городов:")
print(north_stats)

print("\nОписательные статистики для южных городов:")
print(south_stats)

# Вычисляем средние значения смертности для северных и южных городов
mean_mortality_north = north_cities['mortality'].mean()
mean_mortality_south = south_cities['mortality'].mean()

if mean_mortality_north > mean_mortality_south:
  print('Средняя смертность в северных городах больше.')
if mean_mortality_north < mean_mortality_south:
  print('Средняя смертность в южных городах больше.')


# Вычисляем средние значения концентрации кальция для северных и южных городов
mean_hardness_north = north_cities['hardness'].mean()
mean_hardness_south = south_cities['hardness'].mean()

if mean_hardness_north > mean_hardness_south:
  print('Средняя концентрации кальция в северных городах больше.')
if mean_hardness_north < mean_hardness_south:
  print('Средняя концентрации кальция в южных городах больше.')
```

2) Постройте 95% доверительные интервалы для средней годовой смертности
по всем южным и северным городам. Отличаются ли границы интервалов?

```python
# Вычисляем доверительный интервал для северных городов
confidence_interval_north = stats.t.interval(0.95, len(north_cities)-1,
                                             loc=north_cities['mortality'].mean(),
                                             scale=stats.sem(north_cities['mortality']))

# Вычисляем доверительный интервал для южных городов
confidence_interval_south = stats.t.interval(0.95, len(south_cities)-1,
                                             loc=south_cities['mortality'].mean(),
                                             scale=stats.sem(south_cities['mortality']))

print("95% доверительный интервал для средней годовой смертности в северных городах:", confidence_interval_north)
print("95% доверительный интервал для средней годовой смертности в южных городах:", confidence_interval_south)
if confidence_interval_north != confidence_interval_south:
  print('Границы отличаются.')
else:
  print('Границы одинаковы.')
```

3) Постройте 95% доверительные интервалы для средней концентрации
кальция в питьевой воде для южных и северных городов. Отличаются ли
границы интервалов?

```python
# Вычисляем доверительный интервал для концентрации кальция в северных городах
confidence_interval_north_hardness = stats.t.interval(0.95, len(north_cities)-1,
                                                     loc=north_cities['hardness'].mean(),
                                                     scale=stats.sem(north_cities['hardness']))

# Вычисляем доверительный интервал для концентрации кальция в южных городах
confidence_interval_south_hardness = stats.t.interval(0.95, len(south_cities)-1,
                                                     loc=south_cities['hardness'].mean(),
                                                     scale=stats.sem(south_cities['hardness']))

print("95% доверительный интервал для концентрации кальция в северных городах:", confidence_interval_north_hardness)
print("95% доверительный интервал для концентрации кальция в южных городах:", confidence_interval_south_hardness)
if confidence_interval_north_hardness != confidence_interval_south_hardness:
  print('Границы отличаются.')
else:
  print('Границы одинаковы.')
```
