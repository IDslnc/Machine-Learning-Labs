(1) Выполните масштабирование данных.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("diabetes.csv")
data
```

```python
X = data.drop('Outcome', axis=1)
y = data['Outcome']

standard = StandardScaler()
standard.fit(X)
standard_x = standard.transform(X)

X_train, X_test, y_train, y_test = train_test_split(standard_x, y, test_size=0.3, random_state=42)
```

(2) Для построения ансамбля в качестве базового метода используйте поочередно изученные ранее методы с наилучшими гиперпараметрами.

```python
base_estimator_svc = SVC(kernel='linear', C = 1)

bagging_model_svc = BaggingClassifier(estimator=base_estimator_svc)
bagging_model_svc.fit(X_train, y_train)
y_pred_svc = bagging_model_svc.predict(X_test)
print("Classification Report для SVC:")
print(classification_report(y_test, y_pred_svc))
print(bagging_model_svc.estimators_)
```

```python
base_estimator_knn = KNeighborsClassifier(n_neighbors=29)

bagging_model_knn = BaggingClassifier(estimator=base_estimator_knn)
bagging_model_knn.fit(X_train, y_train)
y_pred_svc = bagging_model_knn.predict(X_test)
print("Classification Report для KNN:")
print(classification_report(y_test, y_pred_svc))
print(bagging_model_knn.estimators_)
```

```python
base_estimator_tree = DecisionTreeClassifier(max_depth=4)

bagging_model_tree = BaggingClassifier(estimator=base_estimator_tree)
bagging_model_tree.fit(X_train, y_train)
y_pred_svc = bagging_model_tree.predict(X_test)
print("Classification Report для DecisionTree:")
print(classification_report(y_test, y_pred_svc))
print(bagging_model_tree.estimators_)
```

(3) Получите структуру построенной модели.

(4) Возьмите за основу один из базовых алгоритмов и определите оптимальное количество базовых алгоритмов в ансамбле (GridSearchCV).

```python
base_estimator = KNeighborsClassifier(n_neighbors=29) #если я правильно понял, можно выбрать один любой, то я выбирал метод ближ. соседей

bagging_model = BaggingClassifier(estimator=base_estimator)

param_grid = {'n_estimators': [10, 20, 30, 40, 50, 75, 90, 100, 125, 150, 175, 200]}  # Подбор количества базовых алгоритмов

grid_search = GridSearchCV(bagging_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Лучшее количество базовых алгоритмов KNN:", grid_search.best_params_['n_estimators'])
```

```python
base_estimator_knn = KNeighborsClassifier(n_neighbors=29)

bagging_model_knn = BaggingClassifier(estimator=base_estimator_knn, n_estimators=20)
bagging_model_knn.fit(X_train, y_train)
y_pred_svc = bagging_model_knn.predict(X_test)
print("Classification Report для KNN (20 алгоритмов):")
print(classification_report(y_test, y_pred_svc))
print(bagging_model_knn.estimators_)
```

Данная модель чуть хуже предсказывает здоровых, но лучше предсказывает больных.
