5. Поэкспериментируйте с кусочно-линейными преобразованиями

```python
# Normalize the image data to [0, 1]
img1_normalized = img1.astype(np.float32) / 255.0
img2_normalized = img2.astype(np.float32) / 255.0

#преобразования
log_img1 = np.log(1 + img1_normalized)
log_img2 = np.log(1 + img2_normalized)

gamma = 3
power_img1 = np.power(img1_normalized, gamma)
power_img2 = np.power(img2_normalized, gamma)

#показуем картинки и гистограмки
plt.figure(figsize=(15, 10))

plt.subplot(4, 4, 1)
plt.imshow(img1_normalized, cmap='gray')
plt.title('Original Image 1')
plt.subplot(4, 4, 5)
plt.hist(img1_normalized.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 2)
plt.imshow(log_img1, cmap='gray')
plt.title('Log-transformed Image 1')
plt.subplot(4, 4, 6)
plt.hist(log_img1.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 3)
plt.imshow(img2_normalized, cmap='gray')
plt.title('Original Image 2')
plt.subplot(4, 4, 7)
plt.hist(img2_normalized.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 4)
plt.imshow(log_img2, cmap='gray')
plt.title('Log-transformed Image 2')
plt.subplot(4, 4, 8)
plt.hist(log_img2.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 9)
plt.imshow(img1_normalized, cmap='gray')
plt.title('Original Image 1')
plt.subplot(4, 4, 13)
plt.hist(img1_normalized.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 10)
plt.imshow(power_img1, cmap='gray')
plt.title(f'Power-transformed Image 1 (Gamma = {gamma})')
plt.subplot(4, 4, 14)
plt.hist(power_img1.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 11)
plt.imshow(img2_normalized, cmap='gray')
plt.title('Original Image 2')
plt.subplot(4, 4, 15)
plt.hist(img2_normalized.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(4, 4, 12)
plt.imshow(power_img2, cmap='gray')
plt.title(f'Power-transformed Image 2 (Gamma = {gamma})')
plt.subplot(4, 4, 16)
plt.hist(power_img2.ravel(), bins=256, color='blue', alpha=0.7)

plt.tight_layout()
plt.show()

```

6. Попробуйте выполнить пороговую сегментацию для многопиковых
гистограмм, локализуйте отдельные объекты на изображениях с ее помощью.

```python
import cv2

# Применяем глобальную бинаризацию с порогом 127
_, thresh_img1 = cv2.threshold(img1.astype(np.uint8), 110, 255, cv2.THRESH_BINARY)
_, thresh_img2 = cv2.threshold(img2.astype(np.uint8), 132, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(thresh_img1, cmap='gray')
plt.title('Thresholded Image 1')
plt.subplot(2, 2, 3)
plt.hist(thresh_img1.ravel(), bins=256, color='blue', alpha=0.7)

plt.subplot(2, 2, 2)
plt.imshow(thresh_img2, cmap='gray')
plt.title('Thresholded Image 2')
plt.subplot(2, 2, 4)
plt.hist(thresh_img1.ravel(), bins=256, color='blue', alpha=0.7)

plt.show()
```
